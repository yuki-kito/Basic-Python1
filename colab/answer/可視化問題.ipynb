## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

問1.png

```python
url = "https://raw.githubusercontent.com/yuki-kito/Basic-Python1/master/colab/quizzes/matplotlib_quiz/hist_data.csv"
data = pd.read_csv(url)
x = data['aデータ']
y = data['bデータ']
plt.hist(x,color = 'orange',alpha = 0.75,label = "aデータ",bins = 34)
plt.hist(y,color = 'lightblue',alpha = 0.75, label = "bデータ",bins = 34)
plt.title('aデータとbデータの温度に関する度数分布')
plt.xlabel('温度(℃)')
plt.ylabel('度数')
plt.xlim(0,52)
plt.xticks(np.arange(10,60,10))
plt.ylim(0,27)
plt.yticks(np.arange(0,30,5))

plt.legend()
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

問2.png

```python
url1 = "https://raw.githubusercontent.com/yuki-kito/Basic-Python1/master/colab/quizzes/matplotlib_quiz/scatter_data.csv"
data1 = pd.read_csv(url1)
plot1 = data1[data1['class'] == 0]
plot2 = data1[data1['class'] == 1]
plot3 = data1[data1['class'] == 2]

plot1x = plot1['length (cm)']
plot1y = plot1['width (cm)']
plot2x = plot2['length (cm)']
plot2y = plot2['width (cm)']
plot3x = plot3['length (cm)']
plot3y = plot3['width (cm)']
plt.scatter(plot1x,plot1y,label = 'class0',color = 'red')
plt.scatter(plot2x,plot2y,label = 'class1',color = 'blue')
plt.scatter(plot3x,plot3y,label = 'class2',color = 'green')
plt.title('クラス0～2のlengthとwidthに関する散布図')
plt.ylabel('width(cm)')
plt.xlabel('length(cm)')
plt.xlim(4.1,8.1)
plt.xticks(np.arange(4.5,8.5,0.5))
plt.ylim(1.9,4.53)
plt.yticks(np.arange(2.0,5.0,0.5))
plt.grid()
plt.legend()
plt.show()
```

```python
url2 = 'https://raw.githubusercontent.com/yuki-kito/Basic-Python1/master/colab/quizzes/matplotlib_quiz/line_data.csv'
data2 = pd.read_csv(url2)

x = data2['期間']
height1 = data2['件数']
plt.plot(x,height1,color = 'blue',marker = 'o')
plt.title('期間ごとの件数の推移')
plt.xlabel('期間')
plt.ylabel('件数')
plt.grid()
plt.xticks(x,rotation= 30)
plt.show()

print()
print()

height = data2['費用']
plt.bar(x,height)
plt.title('期間ごとの費用の推移')
plt.xlabel('期間')
plt.ylabel('費用(円)')
plt.grid()
plt.xticks(x,rotation= 30)
plt.show()

```

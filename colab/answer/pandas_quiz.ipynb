```python
df_receipt_only_member = df_receipt[~df_receipt["customer_id"].str.startswith("Z")]
df_receipt2 = df_receipt_only_member.groupby('customer_id')['amount'].apply(lambda x:np.sum(x))
df_receipt2 = df_receipt2.fillna(0)
df_receipt2 = df_receipt2.reset_index()
df_receipt2.columns = ['customer_id','amount']
df_receipt2['log_amount'] = np.log10(df_receipt2['amount'])
df_receipt2.head()
```
